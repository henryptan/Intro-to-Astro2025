```python
from astropy.io import fits
from google.colab import drive
drive.mount('/content/drive') #importing Google drive to obtain the fits file
```

```python
#path to image is recorded in names that are easier to recognize/type
file_dir = '/content/drive/MyDrive/'#fits file found online of Messier 51 and imported to Drive
img_name = 'M51_blue.fit' #there are red, blue, green, and luminosity: I'm just using blue
img = fits.getdata(file_dir+img_name)
```

```python
fig = plt.figure(figsize=(8,8)) #creating space for a figure

image = plt.imshow(img, origin = 'lower', vmin=0.019, vmax=0.055) #values adjusted from original python due to different images
cbar = fig.colorbar(image,shrink=0.82) #cbar configures the color bar on the side
cbar.set_label('Surface Brightness (arb. units)', rotation =90, fontsize=12)
plt.xlim([60,140])
plt.ylim([60,140])
plt.xlabel(r'$\Delta$ RA [px]', fontsize=12)
plt.ylabel(r'$\Delta$ Dec [px]', fontsize=12)
```

```python
import pandas as pd
```

```python
stellar = pd.read_csv(file_dir+'GaiaSource_000-000-000.csv') #Sample CSV of Gaia

fig = plt.figure(figsize=[10,10])

plt.scatter(stellar['ra'], stellar['dec'], alpha = 0.5, s=5, lw =0.5, ec = 'k', color='darkorchid')

plt.xlim=([0,362])
plt.ylim=([-75, 75])
plt.xlabel('RA [$\degree$]', fontsize = 14)
plt.ylabel('Dec [$\degree$]', fontsize=14)

plt.title('Approximately 218,500 Stars in a Localized Region', fontsize=16)
plt.savefig('218.500GaiaStars.png')
```

```python
#Print Fibonacci Sequence given amount of numbers to print

#CHANGE THIS VARIABLE TO DENOTE HOW MANY NUMBERS TO PRINT
stopNumber = 10

numberList = np.zeros(stopNumber) #creates a list of zeroes to input the fibonacci numbers in
sum = 1 #will be the number that is changed each iteration and put into the list
previousSum = 0 #takes the value of the previous index
index=0 #records each iteration

for number in numberList:
  if index == 0:
    numberList[0] == 0 #special case -- 1st step is required to be a 0
    index += 1
  elif index == 1:
    numberList[1] = sum #special case -- 2nd step is required to be 1 for the addition to start
    index += 1
  else:
    sum = sum + previousSum #adds the 2nd to last fibonacci number the previous to get the new one
    previousSum = numberList[index-1] #changes the previousSum to the next number in the list for the next iteration
    numberList[index] = sum #de
    index +=1

print(numberList)
```
